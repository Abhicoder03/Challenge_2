```python
def split_group(group, split_ratio=0.8):
    n_total = len(group)
    n_train = int(n_total * split_ratio)
    idx = group.index
    train_idx = idx[:n_train].tolist()
    test_idx = idx[n_train:].tolist()
    return train_idx, test_idx
```

```python
grouped =train_df.groupby('msno')
indices = grouped.apply(split_group)

# concatenate the train and test indices for each group
train_idx = pd.Index([idx for indices in indices for idx in indices[0]])
test_idx = pd.Index([idx for indices in indices for idx in indices[1]])
missing_indices = set(test_idx) - set(train_idx)
print("Indices not present in train_df:", missing_indices)
test_idx = test_idx.difference(missing_indices)
# create the train and test dataframes
train_df = train_df.loc[train_idx]
test_df = train_df.loc[test_idx]
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
train_pt = train_df.pivot_table(index='song_id', columns='msno', values='target')
test_pt = train_df.pivot_table(index='song_id', columns='msno', values='target')

# calculate similarity scores
similarity_scores = cosine_similarity(train_pt.fillna(0))
```

```python
def get_top_k_recommendations(msno, k):
    if msno not in train_pt.columns:
        return []

    # Get the products that the user has not rated
    rated_products = train_pt[msno]
    rated_products = rated_products[rated_products.notna()]
    unrated_products = train_pt.index.difference(rated_products.index)

    predicted_ratings = []
    # Loop over each unrated item
    for song_id in unrated_products:
        # Calculate predicted rating using similarity scores and user ratings
        similarity_scores_for_product = similarity_scores[train_pt.index.get_loc(song_id)]
        user_ratings = train_pt[msno].fillna(0)
        numerator = (user_ratings * similarity_scores_for_product).sum()
        denominator = similarity_scores_for_product.sum()
        if denominator == 0:
            predicted_rating = 0
        else:
            predicted_rating = numerator / denominator

        predicted_ratings.append((song_id, predicted_rating))

    top_product = [product for product, predicted_rating in sorted(predicted_ratings, key=lambda x: x[1], reverse=True)[:k]]
    return top_product
```

```python
top_products=get_top_k_recommendations(1,10)
```

```python
test_pt = test_pt.fillna(test_pt.mean())
```

```python
def evaluate_recommendations(reviewerID, k):
    # get the actual rated products for the user
    actual_rated_products= test_pt[reviewerID]
    actual_rated_products = actual_rated_products[actual_rated_products.notna()]
    actual_rated_products_set = set(actual_rated_products.index)

    # get the top-k recommended products for the user
    top_k_recommendations = get_top_k_recommendations(reviewerID, k)
    top_k_recommendations_set = set(top_k_recommendations)

    # calculate precision and recall
    intersection = actual_rated_products_set.intersection(top_k_recommendations_set)
    if len(top_k_recommendations_set) == 0:
      precision = 0.0
    else:
      precision = len(intersection) / len(top_k_recommendations_set)
    recall = len(intersection) / len(actual_rated_products_set)

    return precision, recall
```

```python
P,R = evaluate_recommendations(1,10)
print("Precision: ",P)
print("Recall: ",R)
```
