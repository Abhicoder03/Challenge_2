```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(members['bd'], kde=True)
plt.title('Distribution of Age (bd)')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.xlim(0, 80)
plt.show()
```

Most of the users age between 20 to 40.

```python
gender_counts = members['gender'].value_counts()

# Plot the gender distribution using a bar plot
plt.figure(figsize=(8, 6))
sns.barplot(x=gender_counts.index, y=gender_counts.values)
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

male and female users are almost in equal amount.

```python
city_counts = members['city'].value_counts()

# Plot the city distribution using a bar plot
plt.figure(figsize=(12, 6))
sns.barplot(x=city_counts.index, y=city_counts.values, palette="viridis")
plt.title('City Distribution')
plt.xlabel('City')
plt.ylabel('Count')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```

Most of the users are from city 1.

```python
registration_counts = members['registered_via'].value_counts()

# Plot the registration method distribution using a bar plot
plt.figure(figsize=(8, 6))
sns.barplot(x=registration_counts.index, y=registration_counts.values, palette="coolwarm")
plt.title('Registration Method Distribution')
plt.xlabel('Registration Method')
plt.ylabel('Count')
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```

Most of the users have registered via 4,7 and 9 method.

```python
# Convert 'registration_init_time' column to datetime
members['registration_init_time'] = pd.to_datetime(members['registration_init_time'], format='%Y%m%d')

# Extract year and month from the registration dates
members['registration_year'] = members['registration_init_time'].dt.year
members['registration_month'] = members['registration_init_time'].dt.month


# Plot the distribution of registration years using a histogram
plt.figure(figsize=(10, 6))
plt.hist(members['registration_year'], bins=range(members['registration_year'].min(), members['registration_year'].max() + 1), edgecolor='black')
plt.title('Distribution of Registration Years')
plt.xlabel('Registration Year')
plt.ylabel('Number of Registrations')
plt.xticks(range(members['registration_year'].min(), members['registration_year'].max() + 1), rotation=45)
plt.show()
```

This clearly shows registration have increased over the years.

```python
import pandas as pd
import matplotlib.pyplot as plt



# Convert 'expiration_date' column to datetime
members['expiration_date'] = pd.to_datetime(members['expiration_date'], format='%Y%m%d')

# Extract year and month from the expiration dates
members['expiration_year'] = members['expiration_date'].dt.year
members['expiration_month'] = members['expiration_date'].dt.month


# Plot the distribution of expiration years using a histogram
plt.figure(figsize=(10, 6))
plt.hist(members['expiration_year'], bins=range(members['expiration_year'].min(), members['expiration_year'].max() + 1), edgecolor='black')
plt.title('Distribution of Expiration Years')
plt.xlabel('Expiration Year')
plt.ylabel('Number of Expirations')
plt.xticks(range(members['expiration_year'].min(), members['expiration_year'].max() + 1), rotation=45)
plt.show()

```

We can see that expiration has peaked in 2017 thus we can also infer that as registraion peaked in 2017-2018 it was due the old users account expiration.

```python
import pandas as pd


# Specify the numerical features you want to include in the correlation matrix
numerical_features = ['bd', 'registered_via', 'city', 'registration_year']

# Compute the correlation matrix
correlation_matrix = members[numerical_features].corr()

# Print the correlation matrix
print("Correlation Matrix:")
print(correlation_matrix)
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", annot_kws={"size": 12})
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```

```python
print(songs.describe())
```

```python
# Distribution of song lengths
plt.figure(figsize=(10, 6))
sns.histplot(songs['song_length'] / 60000, kde=True, bins=50)  # Convert song length to minutes for better readability
plt.title('Distribution of Song Lengths')
plt.xlabel('Song Length (minutes)')
plt.ylabel('Frequency')
plt.xlim(0,20)
plt.show()
```

Follows normal distribution where mean is around 3.75

```python
# Top 10 most frequent genre categories
top_genres = songs['genre_ids'].value_counts().head(10)
plt.figure(figsize=(10, 6))
top_genres.plot(kind='bar', color='skyblue')
plt.title('Top 10 Most Frequent Genre Categories')
plt.xlabel('Genre Category')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()
```

'465' is the most listened genre.

```python
# Top 10 artists with the most songs
top_artists = songs['artist_name'].value_counts().head(10)
plt.figure(figsize=(10, 6))
top_artists.plot(kind='bar', color='lightgreen')
plt.title('Top 10 Artists with the Most Songs')
plt.xlabel('Artist Name')
plt.ylabel('Number of Songs')
plt.xticks(rotation=45)
plt.show()
```

We can infer that songs with various artists are more preffered.

```python
# Top 10 composers with the most songs
top_composers = songs['composer'].value_counts().head(10)
plt.figure(figsize=(10, 6))
top_composers.plot(kind='bar', color='salmon')
plt.title('Top 10 Composers with the Most Songs')
plt.xlabel('Composer')
plt.ylabel('Number of Songs')
plt.xticks(rotation=45)
plt.show()
```

'Neuromancer' is the top composer followed by JOhann and Wolfgang.

```python
# Top 10 lyricists with the most songs
top_lyricists = songs['lyricist'].value_counts().head(10)
plt.figure(figsize=(10, 6))
top_lyricists.plot(kind='bar', color='purple')
plt.title('Top 10 Lyricists with the Most Songs')
plt.xlabel('Lyricist')
plt.ylabel('Number of Songs')
plt.xticks(rotation=45)
plt.show()
```

```python
# Distribution of languages
plt.figure(figsize=(10, 6))
sns.countplot(data=songs, x='language', palette='Set2')
plt.title('Distribution of Songs by Language')
plt.xlabel('Language')
plt.ylabel('Number of Songs')
plt.xticks(rotation=45)
plt.show()
```

language index 52 belongs to more listened songs.
